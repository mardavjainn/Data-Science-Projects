## 1. Strings

This section covers string methods and slicing operations.

```python
overview = "  A thief who steals corporate secrets through dream-sharing technology. "

print("── String Methods ──")
print(f"  Original     : '{overview}'")
print(f"  .strip()     : '{overview.strip()}'")
print(f"  .lower()     : '{overview.lower().strip()[:40]}…'")
print(f"  .upper()     : '{overview.upper().strip()[:40]}…'")
print(f"  .title()     : '{overview.title().strip()[:40]}…'")
print(f"  .replace()   : '{overview.replace('thief','hacker').strip()[:40]}…'")
print(f"  .split()     : {overview.strip().split()[:5]} …")
print(f"  .find('thief'): {overview.find('thief')}")
print(f"  .count('a')  : {overview.count('a')}")
print(f"  len()        : {len(overview.strip())} characters")
print(f"  .startswith(): {overview.strip().startswith('A')}")
print(f"  .endswith()  : {overview.strip().endswith('.')}")

s = overview.strip()
print("\n── Slicing ──")
print(f"  s[0:10]   : '{s[0:10]}'")
print(f"  s[-10:]   : '{s[-10:]}'")
print(f"  s[::2]    : '{s[::2][:20]}…'")
print(f"  s[::-1]   : '{s[::-1][:20]}…'  (reversed)")
```

## 2. Lists

This section covers list operations, methods, and aggregation functions.

```python
top_movies = ["Inception", "Interstellar", "The Dark Knight", "Avatar", "Titanic"]
ratings    = [8.8, 8.6, 9.0, 7.9, 7.8]

print("── List Operations ──")
print(f"  Original        : {top_movies}")
print(f"  Length          : {len(top_movies)}")
print(f"  Index [2]       : {top_movies[2]}")
print(f"  Slice [1:4]     : {top_movies[1:4]}")
print(f"  Reverse slice   : {top_movies[::-1]}")

top_movies.append("Parasite")      ; print(f"  After append    : {top_movies}")
top_movies.insert(2, "Joker")      ; print(f"  After insert[2] : {top_movies}")
top_movies.remove("Avatar")        ; print(f"  After remove    : {top_movies}")
popped = top_movies.pop()          ; print(f"  After pop()     : {top_movies} | popped: {popped}")
top_movies.sort()                  ; print(f"  After sort()    : {top_movies}")
top_movies.reverse()               ; print(f"  After reverse() : {top_movies}")

print(f"\n  min(ratings)  : {min(ratings)}")
print(f"  max(ratings)  : {max(ratings)}")
print(f"  sum(ratings)  : {sum(ratings):.1f}")
print(f"  sorted()      : {sorted(ratings, reverse=True)}")
```

## 3. Dictionaries, Tuples, & Sets

This section covers dictionaries, tuples, set operations, and unpacking.

```python
movie_info = {
    'title':   'Inception',
    'year':    2010,
    'rating':  8.8,
    'genres':  ['Action', 'Sci-Fi', 'Thriller'],
    'revenue': 836_836_967,
}
print("── Dictionary ──")
print(f"  Keys         : {list(movie_info.keys())}")
print(f"  Values       : {list(movie_info.values())}")
print(f"  Get rating   : {movie_info.get('rating', 'N/A')}")
movie_info['director'] = 'Christopher Nolan'
print(f"  After update : {movie_info}")
del movie_info['revenue']
print(f"  After del    : {list(movie_info.keys())}")

print("\n── Tuple ──")
movie_tuple = ('Inception', 2010, 8.8)
title_t, year_t, rating_t = movie_tuple
print(f"  Tuple        : {movie_tuple}")
print(f"  Unpacked     : title={title_t}, year={year_t}, rating={rating_t}")
print(f"  Count 'Inception': {movie_tuple.count('Inception')}")
print(f"  Index of 2010    : {movie_tuple.index(2010)}")

print("\n── Set ──")
genres_a = {'Action', 'Drama', 'Thriller', 'Sci-Fi'}
genres_b = {'Comedy', 'Drama', 'Romance', 'Action'}
print(f"  Set A        : {genres_a}")
print(f"  Set B        : {genres_b}")
print(f"  Union        : {genres_a | genres_b}")
print(f"  Intersection : {genres_a & genres_b}")
print(f"  Difference   : {genres_a - genres_b}")
print(f"  Symmetric Diff: {genres_a ^ genres_b}")
```

## 4. Flow Control

This section covers conditionals, loops, and loop control keywords.

```python
sample_ratings = [8.8, 3.2, 6.5, 9.1, 4.8, 7.4, 5.0, 2.9, 7.8, 6.0]

def classify_movie(rating):
    if rating >= 8.0:
        return ' Masterpiece'
    elif rating >= 7.0:
        return ' Great'
    elif rating >= 5.5:
        return ' Average'
    elif rating >= 4.0:
        return ' Below Average'
    else:
        return ' Poor'

print("── Rating Classification ──")
for r in sample_ratings:
    print(f"  Rating {r:.1f} → {classify_movie(r)}")

print("\n── For Loop (stars) ──")
for i in range(1, 6):
    print(f"  {'★' * i} ({i}/5)")

print("\n── While Loop ──")
count = 0
high_rated = []
while count < len(sample_ratings):
    if sample_ratings[count] >= 7.0:
        high_rated.append(sample_ratings[count])
    count += 1
print(f"  High-rated movies (≥7.0): {high_rated}")

print("\n── Pass / Continue / Break ──")
for r in sample_ratings:
    if r < 3.0:   break
    if r < 5.0:   continue
    pass
    print(f"  Processing rating: {r}")
```

## 5. Comprehensions

This section covers list, dictionary, and set comprehensions.

```python
ratings = [8.8, 3.2, 6.5, 9.1, 4.8, 7.4, 5.0, 2.9, 7.8, 6.0]

high_ratings   = [r for r in ratings if r >= 7.0]
scaled_ratings = [round(r * 10, 1) for r in ratings]
labels         = [classify_movie(r) for r in ratings]

print("── List Comprehensions ──")
print(f"  Original ratings : {ratings}")
print(f"  High (≥7.0)      : {high_ratings}")
print(f"  Scaled (/100)    : {scaled_ratings}")
print(f"  Labels           : {labels}")

matrix = [[round(i * j * 0.5, 1) for j in range(1, 4)] for i in range(1, 4)]
print(f"\n  3×3 Matrix       : {matrix}")

movies_list = ['Inception','Interstellar','The Dark Knight','Avatar','Titanic']
r_list      = [8.8, 8.6, 9.0, 7.9, 7.8]

movie_rating_dict = {m: r for m, r in zip(movies_list, r_list)}
top_movies_dict   = {m: r for m, r in movie_rating_dict.items() if r >= 8.5}

print("\n── Dictionary Comprehensions ──")
print(f"  All    : {movie_rating_dict}")
print(f"  Top    : {top_movies_dict}")

genre_string = "action comedy drama action thriller drama sci-fi comedy"
unique_genres = {g.capitalize() for g in genre_string.split()}
print(f"\n── Set Comprehension ──")
print(f"  Unique genres: {unique_genres}")
```

## 6. Functions

This section covers functions, scope, default arguments, *args, and **kwargs.

```python
def get_movie_grade(rating: float) -> str:
    if   rating >= 8.5: return 'A+'
    elif rating >= 7.5: return 'A'
    elif rating >= 6.5: return 'B'
    elif rating >= 5.0: return 'C'
    else:               return 'D'

def format_revenue(amount: float, currency: str = 'USD') -> str:
    if amount >= 1e9:
        return f"{currency} {amount/1e9:.2f}B"
    elif amount >= 1e6:
        return f"{currency} {amount/1e6:.1f}M"
    else:
        return f"{currency} {amount:,.0f}"

GLOBAL_THRESHOLD = 7.0
def is_recommended(rating):
    local_note = "Based on TMDB ratings"
    return rating >= GLOBAL_THRESHOLD, local_note

def movie_summary(*titles):
    print(f"  Processing {len(titles)} movies:")
    for i, t in enumerate(titles, 1):
        print(f"    {i}. {t}")

def create_movie_record(**kwargs):
    return {k: v for k, v in kwargs.items()}

print("── Scope & Basic Functions ──")
for r in [9.0, 7.5, 5.5, 3.2]:
    rec, note = is_recommended(r)
    print(f"  Rating {r}: Grade={get_movie_grade(r)}  Recommended={rec}  [{note}]")

print(f"\n  Revenue 836836967  : {format_revenue(836836967)}")

print("\n── *args & **kwargs ──")
movie_summary('Inception', 'Avatar', 'The Dark Knight', 'Titanic')
rec = create_movie_record(title='Interstellar', year=2014, rating=8.6, genre='Sci-Fi')
print(f"  Record: {rec}")
```

## 7. Recursion

This section covers recursive functions.

```python
def count_high_rated(ratings_list, threshold=7.0, idx=0, count=0):
    if idx == len(ratings_list):
        return count
    new_count = count + (1 if ratings_list[idx] >= threshold else 0)
    return count_high_rated(ratings_list, threshold, idx + 1, new_count)

sample = [8.8, 3.2, 6.5, 9.1, 4.8, 7.4, 5.0, 7.8, 6.0, 8.2]
result = count_high_rated(sample, threshold=7.0)
print(f"Recursive count of movies ≥ 7.0 in {sample}: {result}")
```

## 8. Functional Programming

This section covers lambda functions, map, filter, reduce, zip, and enumerate.

```python
from functools import reduce

movies = ['inception', 'avatar', 'the dark knight', 'titanic', 'interstellar']
ratings = [8.8, 7.9, 9.0, 7.8, 8.6]
years   = [2010, 2009, 2008, 1997, 2014]

to_title     = lambda s: s.title()
scale_rating = lambda r: round(r / 10 * 100, 1)

titled = list(map(to_title, movies))
scaled = list(map(scale_rating, ratings))
print("── Map ──")
print(f"  titles  : {titled}")
print(f"  scaled  : {scaled}")

high   = list(filter(lambda r: r >= 8.5, ratings))
modern = list(filter(lambda y: y >= 2009, years))
print("\n── Filter ──")
print(f"  ratings ≥ 8.5 : {high}")
print(f"  years ≥ 2009  : {modern}")

total_rating = reduce(lambda acc, r: acc + r, ratings)
avg_rating_  = total_rating / len(ratings)
print("\n── Reduce ──")
print(f"  Sum of ratings : {total_rating:.1f}")
print(f"  Average rating : {avg_rating_:.2f}")

combined = list(zip(titled, ratings, years))
print("\n── Zip ──")
for name, r, y in combined:
    print(f"  {name:<20} {r:.1f}  ({y})")

print("\n── Enumerate ──")
for rank, (name, r, y) in enumerate(sorted(combined, key=lambda x: x[1], reverse=True), 1):
    print(f"  #{rank}  {name:<20}  {r}  ({y})")
```
