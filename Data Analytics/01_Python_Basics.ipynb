## 1. Variables & Data Types

This section introduces basic Python variables and data types.

```python
total_movies = 930000
avg_rating = 6.75
dataset_name = "TMDB Movies Dataset 2023"
is_large_dataset = True
missing_value = None
z = 3 + 4j

for var_name, val in [
    ('total_movies', total_movies),
    ('avg_rating',   avg_rating),
    ('dataset_name', dataset_name),
    ('is_large_dataset', is_large_dataset),
    ('missing_value', missing_value),
    ('complex z',    z),
]:
    print(f"  {var_name:20s} = {str(val):20s}  | type: {type(val).__name__}")
```

## 2. Math Operators

This section demonstrates basic arithmetic, comparison, logical, and bitwise operators.

```python
a, b = 15, 4

print("── Arithmetic ──")
print(f"  {a} + {b}  = {a + b}")
print(f"  {a} - {b}  = {a - b}")
print(f"  {a} * {b}  = {a * b}")
print(f"  {a} / {b}  = {a / b}")
print(f"  {a} // {b} = {a // b}")
print(f"  {a} % {b}  = {a % b}")
print(f"  {a} ** {b} = {a ** b}")

print("\n── Comparison ──")
print(f"  {a} > {b}   → {a > b}")
print(f"  {a} < {b}   → {a < b}")
print(f"  {a} == {b}  → {a == b}")
print(f"  {a} != {b}  → {a != b}")

print("\n── Logical ──")
x, y = True, False
print(f"  True and False → {x and y}")
print(f"  True or  False → {x or  y}")
print(f"  not True       → {not x}")

print("\n── Bitwise ──")
print(f"  5 & 3  = {5 & 3}")
print(f"  5 | 3  = {5 | 3}")
print(f"  5 ^ 3  = {5 ^ 3}")
print(f"  ~5     = {~5}")
print(f"  5 << 1 = {5 << 1}")
print(f"  5 >> 1 = {5 >> 1}")
```

## 3. Print Formatting

This section demonstrates string formatting methods and alignment.

```python
title   = "Inception"
year    = 2010
rating  = 8.8
revenue = 836836967

print(f"f-string      → Movie: {title} ({year}) | Rating: {rating:.1f} | Revenue: ${revenue:,}")
print(".format()     → Movie: {} ({}) | Rating: {:.1f} | Revenue: ${:,}".format(title, year, rating, revenue))
print("%% formatting  → Movie: %s (%d) | Rating: %.1f" % (title, year, rating))

print("\n── Alignment ──")
print(f"  {'Movie':<20} {'Year':>6} {'Rating':>8}")
print(f"  {'-'*20} {'-'*6} {'-'*8}")
for m, y, r in [("Inception",2010,8.8),("The Dark Knight",2008,9.0),("Interstellar",2014,8.6)]:
    print(f"  {m:<20} {y:>6} {r:>8.1f}")
```

## 4. Random Numbers

This section covers random number generation using random and numpy.

```python
import random
import numpy as np

random.seed(42)

print("── random module ──")
print(f"  random float [0,1)   : {random.random():.4f}")
print(f"  random int  [1,10]   : {random.randint(1, 10)}")
print(f"  random float [5,10)  : {random.uniform(5, 10):.4f}")
print(f"  random choice        : {random.choice(['Action','Comedy','Drama','Horror'])}")
print(f"  random sample (3)    : {random.sample(range(100), 3)}")

genres = ['Action','Comedy','Drama','Horror','Thriller','Romance','Sci-Fi','Animation']
random.shuffle(genres)
print(f"  shuffled genres      : {genres}")

print("\n── numpy random ──")
np.random.seed(42)
print(f"  normal distribution  : {np.random.normal(7.0, 1.5, 5).round(2)}")
print(f"  uniform [0,10)       : {np.random.uniform(0, 10, 5).round(2)}")
print(f"  randint [1,1000)     : {np.random.randint(1, 1000, 5)}")
```
